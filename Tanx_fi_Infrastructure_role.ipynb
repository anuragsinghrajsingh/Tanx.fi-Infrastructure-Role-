# Here only delivered data has been considered

# 1) Compute the total revenue generated by the online store for each month in the dataset. 

```python
df.head(10)
```

```python
df['purchase_date'] = pd.DatetimeIndex(df['purchase_date'])
```

```python
revenue_per_month = df.groupby('seller_id').resample('M', on='purchase_date')['price'].sum()
revenue_per_month
```

# 2) Compute the total revenue generated by each product in the dataset

# 'Revenue per product' have to be calculated.

```python
df.head(10)
```

```python
product_revenue = df[df['order_status'] == 'delivered'].groupby('product_id').size()
product_revenue
```

```python
product_revenue = df[df['order_status'] == 'delivered'].groupby('product_id')['price'].sum()
product_revenue
```

```python
product_revenue = df[df['order_status'] == 'delivered'].groupby('product_id')['price'].sum().reset_index()
product_revenue
```

# 3) Compute the total revenue generated by each customer in the dataset.

# 'Revenue per customer' have to be calculated.

```python
df.head(10)
```

```python
revenue_per_customer = df[df['order_status'] == 'delivered'].groupby('customer_id').size()
revenue_per_customer
```

```python
revenue_per_customer = df[df['order_status'] == 'delivered'].groupby('customer_id')['price'].sum()
revenue_per_customer
```

```python
revenue_per_customer = df[df['order_status'] == 'delivered'].groupby('customer_id')['price'].sum().reset_index()
revenue_per_customer
```

# 4) Identify the top 10 customers by revenue generated. 

# 'Revenue per customer' data have to be sorted in descending order.

```python
top_10_customer = df[df['order_status'] == 'delivered'].groupby('customer_id')['price'].sum().reset_index().sort_values("price",ascending=False)
top_10_customer
```

```python
top_10_customer = df[df['order_status'] == 'delivered'].groupby('customer_id')['price'].sum().reset_index().sort_values("price",ascending=False).reset_index()
top_10_customer
```

```python
top_10_customer = df[df['order_status'] == 'delivered'].groupby('customer_id')['price'].sum().reset_index().sort_values("price",ascending=False).reset_index().drop('index',axis = 1)
top_10_customer
```

```python
top_10_customer = df[df['order_status'] == 'delivered'].groupby('customer_id')['price'].sum().reset_index().sort_values("price",ascending=False).reset_index().drop('index',axis = 1)
top_10_customer.head(10)
```
